# WORKING WITH TUPLES

- A tuple is a collection of sequential items
- A tuple is immutable
- A tuple cannot be modified once created. We can't add, remove, change, reverse, sort a tuple.
- Syntax:

```
(item1, item2, item3, item4, ...)
```

### Creating Tuples

````python
# Examples
courses = ("Python", "Power BI", "SQL", "Excel", "Machine Learning") # 5 times
cities = ()     # empty tuple: not useful as we can't modify
id = (1005,)    # create tuple with one item: to do that give comma after the item, otherwise it is considered parentheses
````

### Accessing the items of a tuple
- Same indexing concepts as lists apply here for tuples

````python
courses[0]      # first item
````

````python
courses[1]      # second item
````

````python
courses[-1]
````

### Unpacking item of a tuple
- In fact, this applies equally to the list object as well

````python
courses
````

````python
# unpacking a tuple
c1, c2, c3, c4, c5 = courses
````

````python
print(c1, c2, c3, c4, c5, sep="\n")
````

````python
# try to change a tuple item
courses[-1] = "AI"
````

````python
courses.index("Machine Learning")
````
